### It is recommended, that when plotting a list of numbers, then using numpy arrays is the best option, because, then you can edit or manipulate your array.

```python
plt.axis([0,5,0,20]) #[xmin, xmax, ymin, ymax]
plt.title("My first Plot")
x = np.array([1,2,3,4])
plt.plot(x,x**2,'ro') # r -> red color, o -> circle points.
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(["Square of x ($x^2$)"],loc = 2) #using $...$ to put latex text.
plt.grid(True)
```

```python
#Plotting sin(x), cos(x), tan(x).
x = np.linspace(-2*np.pi,2*np.pi,50)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
plt.title("Mixed plot")
plt.plot(x,y1,'r-.',x,y2,'b--',x,y3,'c-')
plt.axis([-2*np.pi,2*np.pi,-1.5,2.5])
vec = ['sin(x)','cos(x)','tan(x)']
plt.legend(vec)
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
# Adding a text to the plot.
x = np.array([1,2,3,4,5])
y = x**2
plt.plot(x,y,'bo:')
plt.axis([0,6,0,30])
plt.grid(True)
plt.text(1,2,"$1^2$")
plt.text(2,5,"$2^2$")
plt.text(2.8,10,"$3^2$")
plt.text(3.8,17,"$4^2$")
plt.text(5,26,"$5^2$")
plt.title("Plot of $x^2$")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(["Just Adding text to this plot"],loc = 2)
```

### Locations for the legend.

<li> loc = 0 $\rightarrow$ best 
<li> loc = 1 $\rightarrow$ upper right
<li> loc = 2 $\rightarrow$ upper left
<li> loc = 3 $\rightarrow$ lower right
<li> loc = 4 $\rightarrow$ lower left
<li> loc = 5 $\rightarrow$ right
<li> loc = 6 $\rightarrow$ center left
<li> loc = 7 $\rightarrow$ center right
<li> loc = 8 $\rightarrow$ lower center
<li> loc = 9 $\rightarrow$ upper center
<li> loc = 10 $\rightarrow$ center    

```python
plt.axis([0,4.5,0,100])
plt.grid(True)
plt.title("Mix plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
x = np.array([0,1,2,3,4])
y1 = x**2
y2 = 1.5*x**3
y3 = x**2.5
y4 = np.array([0,30,50,60,65])
plt.plot(x,y1,'ro:') #adding circles using (o)
plt.plot(x,y2,'bs--') #adding stars using(*)
plt.plot(x,y3,'g*-') #adding squares using(s)
plt.plot(x,y4,'m^-.') #adding triangles using (^)
vec = ["y = $x^2$",'y = 1.5$x^3$','y = $x^{2.5}$',r'y = $Random$']
plt.legend(vec,loc = 2)
```

### <u>Color Codes</u>
<p><i> b: blue, g: green, r: red, c: cyan, m: magenta, y: yellow, k: black, w: white <i></p>
